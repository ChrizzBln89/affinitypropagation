```python
import yfinance as yf
import pandas as pd
import tqdm

def quotes_update():
    data = pd.read_csv('/home/chris/code/affinitypropagation/data/Yahoo Ticker Symbols - September 2017.csv')
    tickers = list(data['Yahoo Stock Tickers'].dropna().iloc[2:].values)[0:10]

    dfs = {}

    for symbol in tqdm.tqdm(
    tickers,
    ncols=100,
    desc="Progress merge: ",
    colour="#00d111",):
        try:
            ticker = yf.Ticker(symbol)
            info_dict = ticker.info
            df = pd.DataFrame.from_dict(info_dict, orient='columns')
            df = df.iloc[0:1] 
            dfs[ticker] = df
        except:
            continue

    dfs = dfs.values()
    df = pd.concat(dfs, axis=0).reset_index()
    df["timestamp"] = pd.Timestamp.now()
    df.columns = [x.replace(' ','_').lower() for x in df.columns]
    return df
```

```python
quotes_update()
```
